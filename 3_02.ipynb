```python
def update_data(li, n):
    li.append(4)
    n += 10

list_a = [1, 2, 3]
val_b= 10

update_data(list_a, val_b)
print(list_a, val_b)
```

# 출력 결과: [1, 2, 3, 4] 10
# 이유 설명: list_a(리스트)는 가변객체로 함수 내부에서 수정하면 원본 데이터가 직접 변함 // val_b(정수)는 불변객체로 함수 내부에서 값을 변경해도 원본은 바뀌지 않고 새로운 객체만 생성됨(함수 밖의 원본 val_b에는 영향이 없다.)
