```python
visitor_data = [("2026-03-01", 120), ("2026-03-02", 150), ("2026-03-03", 90), ("2026-03-04", 200)]


dates, counts = zip(*visitor_data)


dates = list(dates)
counts = list(counts)

average_counts = sum(counts) / len(counts)

print(f"날짜 리스트: {dates}")
print(f"방문자수 리스트: {counts}")
print(f"평균 방문자수: {average_counts}")
```
