```python
from abc import ABC, abstractmethod

class OrderItem:
    def __init__(self, name, unit_price, qty):
        self.name = name
        self.__unit_price = 0
        self.__qty = 0
        self.set_unit_price(unit_price)
        self.set_qty(qty)

    def set_unit_price(self, price):
        if price >= 0: self.__unit_price = price
        else: print("가격/수량은 0 이상이어야 합니다.")

    def set_qty(self, qty):
        if qty >= 0: self.__qty = qty
        else: print("가격/수량은 0 이상이어야 합니다.")

    def total_price(self):
        return self.__unit_price * self.__qty

    def __str__(self):
        return f"{self.name} x{self.__qty} = {self.total_price()}원"

class Payment(ABC):
    @abstractmethod
    def pay(self, amount):
        pass

class CardPayment(Payment):
    def pay(self, amount):
        print(f"카드로 {amount}원 결제 완료")

class CashPayment(Payment):
    def pay(self, amount):
        print(f"현금으로 {amount}원 결제 완료")

class Cart:
    def __init__(self):
        self.items = []

    def add(self, item):
        self.items.append(item)

    def total(self):
        return sum(item.total_price() for item in self.items)

    def checkout(self, payment_method: Payment):
        amount = self.total() 
        payment_method.pay(amount) 


if __name__ == "__main__":
    my_cart = Cart()
    my_cart.add(OrderItem("아메리카노", 3000, 2))
    my_cart.add(OrderItem("치즈케이크", 5500, 1))

    card = CardPayment()
    cash = CashPayment()

    my_cart.checkout(card)  
    my_cart.checkout(cash)  
```
