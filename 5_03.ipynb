```python
file_name = "students.txt"

while True:
    name = input("이름: ")
    if name.lower() == 'exit':
        break

    try:
        score = int(input("점수: "))
        with open(file_name, "a", encoding="utf-8") as f:
            f.write(f"{name}, {score}\n")
    except ValueError:
        print("정수로 입력하세요.")

print("-" * 6)

try:
    with open(file_name, "r", encoding="utf-8") as f:
        lines = f.readlines()
        
    scores = []
    high_scores = []

    for line in lines:
        parts = line.strip().split(',')
        if len(parts) == 2:
            name = parts[0].strip()
            val = parts[1].strip()

            if val.isdigit():
                score_int = int(val)
                scores.append(score_int)

                if score_int >= 90:
                    high_scores.append(f"{name}({score_int})")
            

    if scores:
        average = sum(scores) / len(scores)
        print(f"전체 학생 수: {len(scores)}명")
        print(f"전체 학생 평균: {average:.2f}점")
    else:
        print("데이터 X")

    if high_scores:
        print(f"90점 이상: {', '.join(high_scores)}")
    else:
        print("90점 이상인 학생이 없음")

except FileNotFoundError:
    print("파일 X")
```
