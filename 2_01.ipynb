```python
stored_id = "admin"
stored_pw = "pass123"


while True:
    input_id = input("아이디를 입력하세요: ")
    input_pw = input("비밀번호를 입력하세요: ")
    
   
    if input_id == stored_id and input_pw == stored_pw:
        print("환영합니다, 관리자님!")
        break 
    else:
        print("아이디 또는 비밀번호가 틀렸습니다. 다시 시도하세요.")
```
