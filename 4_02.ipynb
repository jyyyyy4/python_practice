```python
class OrderItem:
    def __init__(self, name, unit_price, qty):
        self.name = name
        self.__unit_price = 0
        self.__qty = 0

        self.set_unit_price(unit_price)
        self.set_qty(qty)

    def set_unit_price(self, price):
        if price >= 0:
            self.__unit_price = price
        else:
            print("가격/수량은 0 이상이어야 합니다.")

    def set_qty(self, qty):
        if qty >= 0:
            self.__qty = qty
        else:
            print("가격/수량은 0 이상이어야 합니다.")

    def get_unit_price(self):
        return self.__unit_price
    
    def get_qty(self):
        return self.__qty
    def total_price(self):
        return self.__unit_price * self.__qty
    def __str__(self):
        return f"{self.name} x{self.__qty} = {self. total_price()}원"
    
if __name__ == "__main__":
    item = OrderItem("아메리카노", 3000, 2)
    print(item) 

    item.set_unit_price(3500)
    item.set_qty(3)
    print(f"변경 후: {item}")

    item.set_qty(-5)
    print(f"최종 유지 값: {item}")        
```
